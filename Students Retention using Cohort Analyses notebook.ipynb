# Data Wrangling (Cleaning)

```python
# Fill in the missing values in online_order with mean
online_order_mode = std_df['online_order'].mode()
std_df.online_order.fillna(online_order_mode, inplace=True)
```

# Alternative approach for online_order

```python
std_df.online_order.value_counts()
```

```python
# Since online_order values are in binary form, we can use (1.0)
std_df.online_order.fillna(1.0, inplace= True)
```

```python
# for brand
brand_mode = std_df['brand'].mode()
std_df.brand.fillna(brand_mode, inplace=True)
```

```python
# for product_line
product_line_mode = std_df['product_line'].mode()
std_df.product_line.fillna(product_line_mode, inplace= True)
```

```python
# For product_class
product_class_mode = std_df['product_class'].mode()
std_df.product_class.fillna(product_class_mode, inplace = True)
```

```python
# for product_size  
product_size_mode = std_df['product_size'].mode()
std_df.product_size.fillna(product_size_mode, inplace= True)
```

```python
# For standard_cost 
standard_cost_mean = round(std_df['standard_cost'].mean(), 0)
std_df.standard_cost.fillna(standard_cost_mean, inplace = True)
```

```python
import datetime as dt
```

```python
# transactionmonth
# cohortmonth
# cohortindex
```

```python
date_extract=lambda x: (x.year, x.month, 1)
```

```python
# Then apply it to the transaction_date
std_df.transaction_date.apply(date_extract)
```

```python
# Convert it to datetime
date_extract=lambda x:dt.datetime(x.year, x.month, 1)
```

```python
std_df.transaction_date
```

```python
std_df['transactionMonth']= std_df.transaction_date.apply(date_extract)
```

```python
std_df
```

```python
# To get the cohortMonth, make a query for a particular customer,
std_df.query('customer_id==2950')
```

```python
# Then use groupby to capture all the transactions made by the customer
std_df['CohortMonth']=std_df.groupby('customer_id')['transactionMonth'].transform(min)
```

```python
std_df.query('customer_id==2950')
```

```python
# Find the cohort index
# Cohort_index = year_diff *12 + month_diff + 1
```

```python
year_diff = std_df['transactionMonth'].dt.year - std_df.CohortMonth.dt.year
month_diff = std_df['transactionMonth'].dt.month - std_df.CohortMonth.dt.month
```

```python
std_df['Cohort_index']= year_diff *12 + month_diff + 1
```

```python
std_df
```

```python
# To get the cohort data,
Cohort_data=std_df.groupby(['Cohort_index', 'CohortMonth'])['customer_id'].apply(pd.Series.nunique).reset_index()
```

# Retention metric

```python
cohort_data_pivot=Cohort_data.pivot_table(columns="Cohort_index", index ="CohortMonth", values = "customer_id")
```

```python
cohort_data_pivot
```

```python
final =round(cohort_data_pivot.divide(dvd, axis = 0)*100, 2)
final.index = final.index.strftime('%Y-%m')
```

```python
# Initialize the figure
plt.figure(figsize=(16, 10))
# Adding a title
plt.title('Average Standard Cost: Monthly Cohort', fontsize = 14)
# Creating the heatmap
sns.heatmap(final, annot = True, vmin = 0.0, vmax = 100, cmap = 'RdYlBu', fmt = 'g')
plt.ylabel('Cohort Month')
plt.xlabel('Cohort Index')
plt.yticks(rotation='360')
plt.show()
```
